### Python is different from other languages:
This is a line break. It is used in most high-level languages to mark a block of code

{

}

In Python, identation is used instead of line breaks (e.g {} or ;)

Because of this, spaces and tabs are important!
so choose only one method when moving onto the next line

```python
#usual line break
'''
if(condition){
hello();
}
'''
#python indentation
'''
if (condition):
    hello()
'''
```

### 2. Print
You can display most results on the console using this command

Each print command automatically prints on a newline

```python
print("HELLO, WORLD!")
```

```python
print("Shrek")
# Output: Shrek

print("Shrek","Two") # Comma appends with a space between each by default!
# Output: Shrek Two

print("Shrek","The Ogre", sep=", ") # You can change this default though!
# Output: Shrek, The Ogre

print("Shrek\nMovie Perfection") # \n is a newline character!

''' Output:
Shrek
Movie Perfection
'''

print("A" * 5) # You can multiply!
# Output: aaaaa

print("Hello\nWorld") #\n newline
print("Hello\tWorld") #\t tab
```

```python
# Advanced: (We will be skipping this during the workshop)

name = 'Caleb'
state = 'cool art'
number = 3

# using % (%d-integer %s-string, %f-float %b-boolean)

print("%d %s %s %s" % (3,"Digital","Design","Development"))

# using .format for strings

print('{} makes {}'.format(name,state))
# Custom index helps you to do assign positions
print('{1} makes {0}'.format(name,state))
# Named arguments
print("{name}".format(name="Caleb"))
# Output: Caleb

# You can even use it to set decimal places
print("{:.2f}".format(number))
    # Output: 3.00
    # Notice the number decimal places is 2!

# More Format options!
# {:<5} Left padding within 5 width(align left)
# {:>5} Right padding within 5 width(align right)
# {:^5} Centre padding within 5 width(align centre)
# {:.5} Truncate string
# {:d} Int 
# {:f} Float

# {:5d} Padded Int (fill empty spaces till width 5)
# {:.5f} Fixed Decimal Float (fixed number of decimals)
# {:08b} Binary with 8 digits zero-padded on the left

# {:+d} Sign Prefixed Number
# Eg: {42:+d} : +42 , {-42:+d} : -42
# {: d} Space Prefixed Number
# {:=5d} Control the position of the sign relative to padding!
# Eg: {42: d} : [space]42 , {-42: d} : -42
```

### 3. Variables:
In python, all variables are considered objects.
In a sense, we can visualise them as containers for types of data (classes)

Naming:
- All names are Case-Sensitive
- They also cannot be keywords used in Python



- if they are highlighted in a color other than black, they cannot be used as variable names
- usually we use the convention:
    - eg. my_name, age_number


4 Main Types:
  - String (text) --> str
  - Integers --> int
  - Float --> float
  - Boolean --> bool


```python
my_string = "Stay Up Till Dawn"
my_decimal = 5.75
my_integer = 6
my_boolean = True

print(type(my_string))
print(type(my_integer))
print(type(my_decimal))
print(type(my_boolean))

# if you want to compare between types:
type(my_string)==str
isinstance(my_string,str)
```

We can also convert variables to different types.

- Bear in mind that not all types can be converted to another

```python
print(int(my_decimal))
print(bin(my_integer)) #converting to binary -base 2
print(bool(my_string)) # if not empty string
```

### 6. Conditionals

Conditionals refer to decisions you have to make during the running of the program.
This allows evaluation of conditions before any consequential action is taken.



- Important to note:
#### " == " is not " = "
    - == compares values while = assigns them

```python
# == equals to
# != not equal to
# >  more than
# <  less than
# >= more than or equal to
# <= less than or equal to

# is 
#Checks if something is the exact same object 

#Eg.
print(1 == True) #Returns True
print(1 is True) #Reurn False as it not the same object
```

```python
print(True and False)
print(True or False)
print(not True)
```

LOGIC Operations


```python
#Therefore if is the command we will be using to check conditionals

grade = 50
if grade >= 50:
    print("pass")
elif grade >= 30:
    print("you can do better")
else:
    print("RIP your GPA")
```

```python
#lets try a bit more complicated now
grade = 50
is_better_than_avg = True

if grade >= 50 and is_better_than_avg:
    print("DAMN BOI")
elif grade >= 50 and not is_better_than_avg:
    print("you can do better fam")
else:
    print("RIP your GPA")
```

```python
# Advanced:

# any(): returns True if 1 of more elements is True
# all(): returns True only if all elements are True

print(all([True, False]))
print(any([True, False]))
```
